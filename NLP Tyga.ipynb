```python
tyga_all_lyrics= tyga.lyrics_cleaned.sum()
```

```python
counter = Counter(tyga_all_lyrics.split())
```

### Top words

```python
counter.most_common(50)
```

## Bigrams and trigrams / words around top words

```python
chars=[]
for lyric in tyga.lyrics:
    for c in lyric:
        if c.isalnum()==False and c!='*' and c!="'" and c!=" ":
            chars.append(c)
```

```python
def clean_2(s):
    r = s.lower().strip()
    for c in chars:
        r = r.replace(c, ' ')
    rs = [w for w in r.split()] # if w not in stop and len(w) > 1]
    r = ' '.join(rs)
    return r
```

```python
top_cleaned_50_words = [w for w, f in counter.most_common(50)]
tokens=tyga_all_lyrics.split()
```

### Bigrams

```python
for token in top_cleaned_50_words:
    print('WORD:', token)
    counter = Counter([ng for ng in ngrams(tokens, 2) if token in ng])
    print(counter.most_common(10))
    print()
```

### Trigrams

```python
for token in top_cleaned_50_words:
    print('WORD:', token)
    counter = Counter([ng for ng in ngrams(tokens, 3) if token in ng])
    print(counter.most_common(10))
    print()
```

## NGRAMS

```python
def find_ngrams(input_list, n):
    return zip(*(input_list[i:] for i in range(n)))
```

```python
bigrams=Counter(find_ngrams(tokens,2))
```

```python
bigrams.most_common(50)
```

```python
trigrams=Counter(find_ngrams(tokens,3))
```

```python
trigrams.most_common(50)
```
